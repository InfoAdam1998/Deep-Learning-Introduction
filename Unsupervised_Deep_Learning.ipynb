# Self Organizing Maps

**How Does a Self-Organizing Map (SOM) Work?**
A Self-Organizing Map (SOM) is an unsupervised learning algorithm that creates a 2D map from high-dimensional input data, helping to visualize patterns or clusters in the data. SOMs are based on the idea that similar data points should be placed close to each other on this 2D map, preserving the topology of the original data space.

**Steps of SOM:**
* Input Data and the Grid: The input data is usually high-dimensional (e.g., a 10D vector representing some features of an image).
The SOM grid (the "map") is a 2D array of neurons (nodes). Each neuron has a weight vector of the same dimension as the input data.
Random Weight Initialization:

* Each neuron on the grid is initialized with a random weight vector of the same dimension as the input data (i.e., the same number of features).
Finding the Best Matching Unit (BMU):

* For each input data point, the SOM calculates the distance (usually Euclidean distance) between the input vector and the weight vectors of all neurons.
The neuron with the weight vector most similar to the input data is called the Best Matching Unit (BMU).
Updating the Weights:

* Once the BMU is identified, the weights of the BMU and its neighboring neurons are adjusted to become more similar to the input data.
The closer the neuron is to the BMU on the grid, the more its weight is updated. This is done using a neighborhood function (like a Gaussian function) and a learning rate that decays over time.
Repeating the Process:

* This process is repeated for many iterations, where each input data point is presented to the SOM multiple times. Over time, the neurons "self-organize" to reflect the structure of the input data.
After training, the neurons in the SOM become tuned to different regions of the input space, forming a topologically ordered map.

**How the Coordinates of a Node Relate to the Input Space:**
* Coordinates on the SOM Grid: Each neuron in the 2D SOM grid has a fixed position, like (i, j) in the grid.
The coordinates on this grid do not have a direct relationship to the original input space's dimensions. Instead, the weight vector associated with each neuron represents a point in the input space.

* Weight Vector in Input Space: The weight vector of each neuron can be thought of as the coordinates of that neuron in the high-dimensional input space.
When the SOM is fully trained, nearby neurons on the 2D grid will have similar weight vectors, meaning that they respond to similar inputs.
The final trained map is a topological representation of the input space, where neighboring neurons represent similar clusters in the input space.

**Example (Simplified):**
* Imagine you have a dataset with three features (3D data points). You want to project this data into a 2D space using SOM.

* You initialize a 2D grid of neurons (e.g., a 5x5 grid). Each neuron has a 3D weight vector (randomly initialized).
* You input a 3D data point [0.5, 0.2, 0.8].
* The SOM finds the neuron whose weight vector is closest to [0.5, 0.2, 0.8] in the 3D input space (BMU).
* The BMU and its neighboring neurons update their 3D weight vectors to be closer to [0.5, 0.2, 0.8].
* This process repeats for other data points, and after training, the 2D grid of neurons will represent clusters of similar data points from the 3D space.

**Key Idea:** Neurons in close proximity on the SOM grid will respond to similar input data points, even though their weight vectors are high-dimensional. SOM preserves the topology of the input space, meaning that data points that are close in the input space will be mapped to nearby neurons on the grid.

**In Summary:** SOM projects high-dimensional input data into a 2D map.
Neurons in the SOM grid are associated with weight vectors in the input space.
The "coordinates" of neurons in the input space are represented by their weight vectors, and nearby neurons on the grid represent similar regions in the input space.

**Common questions**

**How It Knows Its Similar to the Input Data:**
* When an input is presented, the SOM identifies the Best Matching Unit (BMU), which is the node most similar to the input data. To determine this, the SOM calculates the distance (often using Euclidean distance) between the input vector and each nodes weight vector (which represents the nodes current position in the input space). The node with the smallest distance is selected as the BMU.

**Does the Radius Shrink After Each Epoch?**
* Yes, the radius shrinks after each epoch (or iteration). Early in the training, the radius is wide to allow broad generalization, and as the SOM learns, the radius narrows down to focus on refining individual nodes, enhancing the specificity of the patterns it captures.

## SOM Algorithm
* Grid of Neurons: Imagine you have a 5x5 grid of neurons, totaling 25 neurons. Each neuron represents a region in the input space (e.g., different characteristics of customers).

* Calculating Distances: When you input a new observation (like a customer's data), each of the 25 neurons calculates the Euclidean distance between its weight vector (representing that neuron's learned features) and the input vector (the customer's data).

* Finding the Best Matching Unit (BMU): The neuron with the lowest Euclidean distance is identified as the Best Matching Unit (BMU). This means it's the neuron that is most similar to the input observation.

* Updating the BMU and Neighbors: After identifying the BMU, that neuron, along with its neighbors (based on the neighborhood radius), will adjust their weight vectors. This adjustment is influenced by how close they are to the BMU, allowing them to learn from the input data.

* Repeating the Process: This process is repeated for every observation in your dataset. Over time, as more data is fed into the SOM, the neurons will learn to represent different clusters of data points, with similar observations being grouped together.

This iterative process helps the SOM to organize the input data in a way that preserves the relationships and similarities between different observations.

# Boltzman Machines

**Boltzmann Machines Overview**

Boltzmann Machines (BMs) are a type of neural network particularly useful for tasks like pattern recognition and complex decision-making. They operate based on probabilities and utilize concepts from statistical mechanics and energy states to model data.

**Neurons & Energy States**
* Each neuron in a Boltzmann Machine can be in two states: "on" (active) or "off" (inactive), represented as binary values (0 or 1).
Neurons interact through weights, determining the influence one neuron has on another.

* In a Boltzmann Machine, neurons (or nodes) during training can be in a state of 1 (on) or 0 (off), which means they are either active or inactive. This binary behavior is crucial for the way the machine learns patterns in the data.

**Here's how it works:**

* 1 (On): When a neuron is "on" (state 1), it means that neuron is actively contributing to the current configuration or pattern in the network.
* 0 (Off): When a neuron is "off" (state 0), it means that neuron is not contributing to the pattern at that moment.

**Low Energy:**
* If the activated neurons are part of a pattern the model has already learned well, they work smoothly together. The network feels "comfortable" in this configuration, which means it's in a low-energy state. This indicates the neurons are responding in a stable, meaningful way that aligns with the learned patterns.

**High Energy:**
* If a neuron is activated but doesnt fit well with the pattern or its connected to neurons in a way that doesnt reflect the data, the network will feel "uncomfortable," leading to a high-energy state. This happens because the neurons arent cooperating in a way that represents a clear pattern from the data.

Its like the neuron is confused or unsure how to react because the pattern its part of doesnt match what it has seen before or doesnt align with the other neurons' behavior.

**Learning with Probabilities**
* BMs learn by adjusting the weights between neurons to uncover patterns in the data.
* They explore different combinations of neuron states and calculate the energy of each configuration.
The weights are adjusted to increase the likelihood of settling into states that represent training data patterns, with neurons flipping states based on the probability of lowering overall energy.

**Hidden & Visible Neurons**
* Visible Neurons: These represent the data directly (e.g., pixel values in an image).
Hidden Neurons: They capture complex relationships and underlying patterns in the data, enabling the network to recognize features not

**Unique Properties of Boltzmann Machines**
* No Directionality: Unlike traditional neural networks with a clear input-output structure, Boltzmann Machines feature bidirectional connections. This means any neuron can influence any other, allowing the network to model variable dependencies in both directions without assuming a fixed flow of information.
* Symmetric Connections: The connection strength between neurons is the same in both directions.
* Data Generation: After training, Boltzmann Machines can generate new data points by sampling from the learned distribution. This makes them generative models, capable of creating new examples resembling the training data.

For instance, in a movie recommendation system, a trained BM can generate ratings or suggest movies based on learned patterns from previous user ratings. In image generation, it can create new images by sampling from hidden neurons representing different features.
* Input/Output Structure: BMs differ from traditional networks as they lack a strict input-output flow. The visible neurons serve as both inputs and outputs, while hidden neurons capture deeper patterns.

Visible neurons accept training data and later output learned data, while hidden neurons interact with visible ones to identify complex relationships.

**Summary of Key Points**
* No Directionality: BMs have bidirectional connections, lacking fixed input-output layers.
* Generative Model: They can generate new data mimicking training patterns.
* Input/Output Flexibility: The same visible neurons function for both input and output, while hidden neurons capture deeper patterns.
Computational Complexity

Despite their flexibility and power for certain tasks, Boltzmann Machines can be computationally expensive due to the need to calculate probabilities across all neuron pairs. This complexity is one reason why more efficient variants, like **Restricted Boltzmann Machines (RBMs)**, are often preferred in practice. RBMs simplify training by restricting connections between neurons, allowing for faster convergence and reduced computational cost.

#RBM Algorithm

BRM stands for Boltzmann Restricted Machine (BRM), more commonly known as a Restricted Boltzmann Machine (RBM). It is a simplified version of a Boltzmann Machine, with specific restrictions that make it more practical and efficient for certain tasks, particularly in unsupervised learning.

**Simplified Structure:**
* An RBM has two layers: a visible layer (input layer) and a hidden layer.
Unlike a full Boltzmann Machine, there are no connections between neurons within the same layerthis is the "restriction" in RBMs. Neurons in the visible layer only connect to neurons in the hidden layer and vice versa.

**Input Representation:**

* The users ratings (visible neurons) are provided as input. This can include existing ratings for some movies, while others may be left unrated (represented as zeros).

**Activation of Hidden Neurons:**

* The RBM calculates the activation of the hidden neurons based on the input ratings. This involves using the learned weights (connections between visible and hidden neurons) to determine how strongly each hidden neuron should be activated based on the visible inputs.
The activation is typically done using a probabilistic method where each hidden neuron is turned on or off based on the input it receives, effectively capturing underlying patterns and features.

* RBMs aim to minimize the energy function, just like standard Boltzmann Machines. Low-energy configurations represent strong patterns in the input data, while high-energy configurations represent randomness or noise.The goal is to find low-energy patterns by adjusting the weights between the visible and hidden layers.

**Sampling:**

* The RBM generates a sample from the hidden layer. This means it calculates the probabilities of each hidden neuron being active given the current input from the visible layer.
The sampling process allows the RBM to capture the latent factors that contribute to user preferences.

* RBMs are trained using a method called contrastive divergence. This method helps the RBM learn by gradually adjusting the weights to minimize the difference between the input data and the patterns the network generates. It leverages Gibbs sampling, where the network flips neurons between on/off states during training, gradually refining the weights.

**Reconstructing Visible Neurons:**

* Using the activated hidden neurons, the RBM then reconstructs the visible layer (movie ratings). This involves computing the output for each movie based on the active hidden neurons.
The reconstruction process produces predicted ratings for all movies, including those the user hasnt rated yet.

**Output of Predictions:**

* The predicted ratings for the unrated movies can then be used to recommend movies to the user. For instance, the RBM may suggest the top-rated movies based on the reconstructed visible neurons, which represent the users likely preferences.

#Contrastive Divergence

**Contrastive Divergence (CD)** starts by taking the input and making an initial guess about the pattern it's trying to learn.
Gibbs Sampling is a technique that helps refine these guesses by flipping the "on" or "off" state of the neurons. It's not based on normal distribution, but rather its a method that sequentially updates the state of one neuron at a time, keeping others fixed, and gradually refines the learned patterns through this process. This process improves the accuracy of the guessed pattern.

**What is Gibbs Sampling?**
Its a process where each neurons state is updated based on the current states of the other neurons. The system doesnt change everything at once but does it step-by-step, adjusting based on the influence of connected neurons. Over time, this gradual adjustment helps the network better capture patterns from the input data.

**Adjusting Weights Without Affecting Other Neurons**
In RBMs, even though the connections between neurons are undirected, the process of adjusting weights happens through the feedback loop in both directions. Each weight is adjusted based on the correlation between the connected neurons. The balance comes from the entire network gradually learning the dependencies through multiple iterations of training. So, while one weight gets adjusted, the rest of the network adapts as well, maintaining overall learning stability.

## Deep Belief Networks

* Definition: A Deep Belief Network is a type of neural network composed of multiple layers of stochastic, latent variables. Each layer consists of a Restricted Boltzmann Machine (RBM) or a similar generative model.
* Structure: DBNs are typically stacked RBMs, where the output of one RBM serves as the input to the next. This allows for hierarchical learning of features.
* Learning Process: They are trained in a layer-wise fashion, where each RBM is trained individually before the entire network is fine-tuned using supervised learning techniques.
* Directionality: DBNs have a directed structure, moving from the input layer through hidden layers to the output layer.
*  Applications: DBNs are used for tasks like image recognition, speech recognition, and dimensionality reduction.


## Deep Boltzmann Machines

* Definition: A Deep Boltzmann Machine is an extension of the Boltzmann Machine that consists of multiple layers of stochastic binary or continuous units.
* Structure: DBMs can have connections between both visible and hidden layers across multiple levels, allowing for more complex interactions than in DBNs.
* Learning Process: Training is more complex due to the need for joint distributions across multiple layers. Like RBMs, DBMs often use Contrastive Divergence for training.
* Directionality: DBMs are non-directed, meaning that the connections between neurons do not have a specified direction; any neuron can influence any other neuron.
* Applications: DBMs are used in similar contexts as DBNs but can capture more intricate relationships due to their fully connected structure, making them suitable for tasks requiring deeper representations.


**Key Differences**

* Structure: DBNs consist of stacked RBMs with directed connections, while DBMs have a non-directed structure with connections across multiple layers.
* Learning Approach: DBNs are typically trained layer-wise, while DBMs require more complex training due to their non-directed connections.
* Use Cases: DBNs are often favored for applications where a clear hierarchy is beneficial, whereas DBMs can handle more complex relationships due to their richer structure.

* Directed Networks (DBNs): Simpler, easier to train, with one-way information flow. Used when you need a more structured, hierarchical learning approach.
* Undirected Networks (DBMs): More complex, bidirectional connections, harder to train but capable of capturing more intricate patterns across all neurons. Used for tasks that require deeper, more interconnected feature learning or data generation.

#AutoEncoders

An autoencoder takes an input (such as an image) and produces an output that aims to replicate the original input. It transforms the input into a lower-dimensional representation through a mapping function in its hidden layers. The weights can be pre-defined, often set to values like -1 and +1. While softmax can be used in classification tasks to convert output values into probabilities, it is generally not utilized in standard autoencoders; instead, other activation functions like ReLU, Sigmoid, or Tanh are more common. Some configurations may represent biases in the model.

* STEP 1: We start with an array where the lines (the observations) correspond to the users and the columns (the features) correspond to the movies. Each cell (u, i) contains the rating (from 1 to 5, 0 if no rating) of the movie i by the user u.

* STEP 2: The first user goes into the network. The input vector x = (r1, r2, ..., rm) contains all its ratings for all the movies.

* STEP 3: The input vector x is encoded into a vector z of lower dimensions by a mapping function f (e.g., sigmoid):
z = f(Wx + b)
where W is the vector of input weights and b is the bias.

* STEP 4: z is then decoded into the output vector y of the same dimensions as x, aiming to replicate the input vector x.

* STEP 5: The reconstruction error d(x, y) = ||x - y|| is computed. The goal is to minimize it.

* STEP 6: Back-Propagation: from right to left, the error is back-propagated. The weights are updated according to how much they are responsible for the error. The learning rate decides how much we update the weights.

* STEP 7: Repeat Steps 1 to 6 and update the weights after each observation (Reinforcement Learning). Or repeat Steps 1 to 6 but update the weights only after a batch of observations (Batch Learning).

# AutoEncoders Variants

**Overcomplete Hidden Layers**
* Definition: When an autoencoder has more hidden neurons than input neurons, it's referred to as "overcomplete." This means the hidden layer is larger than the input layer.
* Effect: Rather than reducing the dimensionality of the data, overcomplete autoencoders try to learn redundant features, which can help capture finer details in the input. However, the risk here is overfitting, where the autoencoder may simply memorize the input without generalizing well to unseen data. In practice, regularization methods are often applied to control this.

**Sparse Autoencoders**
* Definition: Sparse autoencoders apply a regularization technique to enforce a sparsity constraint on the hidden layer, encouraging only a small fraction of neurons to activate at any given time.
* Effect: By forcing only a subset of hidden neurons to be active, sparse autoencoders are better at learning more interpretable, high-level features from the input data. This sparsity reduces the risk of overfitting and helps the model focus on the most important patterns. This property makes sparse autoencoders useful for feature extraction.

**Contractive Autoencoders**
* Definition: Contractive autoencoders apply a regularization term that penalizes the sensitivity of the hidden representation to small changes in the input. This regularization term is based on the Jacobian matrix of the hidden layer with respect to the input.

* Effect: This constraint forces the autoencoder to learn features that are robust to small perturbations in the input, making the learned representations less sensitive to minor changes in the data. Contractive autoencoders are often used for capturing smooth, invariant features, making them useful in applications like manifold learning and creating more stable representations.

**Denoising Autoencoders**
* Definition: Denoising autoencoders are trained to recover the original, uncorrupted input from a noisy version of the input data. Random noise is deliberately added to the input, and the autoencoder learns to filter it out.
* Effect: This noise-resilient training allows the network to focus on the most essential features of the data, improving its ability to generalize. Denoising autoencoders are commonly used for tasks like removing noise from images or reconstructing incomplete or noisy data.

**Stacked Autoencoders**
* Definition: Stacked autoencoders consist of multiple layers of autoencoders, where each layer's output serves as the input to the next. The layers are trained one by one in a greedy, unsupervised manner.
* Effect: By stacking multiple autoencoders, the network can learn hierarchical representations of the data. Early layers may capture basic features, like edges in images, while deeper layers capture more complex patterns, such as object shapes or textures. This hierarchical feature learning makes stacked autoencoders especially powerful for tasks like unsupervised pretraining in deep networks.

**Deep Autoencoders**
* Definition: Deep autoencoders are autoencoders that include multiple hidden layers on both the encoder and decoder sides. The encoder reduces the dimensionality step-by-step, while the decoder progressively reconstructs the original input.
* Effect: These deeper architectures allow the network to capture more abstract and complex features, making them well-suited for tasks such as high-quality data reconstruction, anomaly detection, or dimensionality reduction in large datasets. The deep architecture also makes them more expressive but requires careful training to avoid overfitting.

**Softmax**: Softmax is used for classification tasks to convert raw output into probabilities across multiple classes. Its not typically used in standard autoencoders.

**Bias**: Yes, biases are additional learnable parameters added to neurons that help shift the activation function, improving learning flexibility.

## Why use AutoEncoders Variants

* Combat Overfitting: Overcomplete autoencoders, with more hidden neurons than input features, risk memorizing inputs instead of learning generalizable features.

* Regularization: Variants like Sparse and Contractive Autoencoders introduce regularization to promote meaningful representations and prevent overfitting.

* Sparse Autoencoders: Encourage fewer active neurons, focusing on important patterns, which reduces reliance on all hidden neurons.

* Contractive Autoencoders: Penalize sensitivity to small input changes, ensuring stability and robustness against noise, aiding tasks like manifold learning.

* Denoising Autoencoders: Train to reconstruct clean inputs from noisy versions, enhancing generalization by focusing on core data patterns rather than noise.

* Stacked Autoencoders: Combine multiple layers to learn complex, hierarchical patterns, beneficial for deep representation in complex data (e.g., images, speech).

* Handling Larger Hidden Layers: Larger layers can lead to overfitting. Regularization techniques like sparsity ensure neurons behave effectively, enhancing efficiency.

* Extract Meaningful Features: Sparse autoencoders focus on essential patterns, while contractive autoencoders ensure invariance to small input changes, improving feature recognition in tasks like clustering.
